# Problem Statement:-

 It is difficult to decide a fair Airbnb price in New York City because prices change based on location, room type, and features.  Without data analysis, it is hard to know which factors affect the price most.This makes pricing confusing for both hosts and guests.

# Goal of the Data:-

The goal of this data is to analyze Airbnb listings in New York City and understand the factors that affect pricing.
It helps identify patterns related to location, room type, availability, and reviews.
The data is used to build predictive models that estimate a fair price for Airbnb listings.

# Explanation of the features in the dataset:-

1. **id** – Unique ID number for each Airbnb listing

2. **name** – Name/title of the Airbnb listing

3. **host_id** – Unique ID of the host

4. **host_name** – Name of the host

5. **neighbourhood_group** – Main area of NYC (Manhattan, Brooklyn, Queens, Bronx, Staten Island)

6. **neighbourhood** – Specific neighborhood where the listing is located

7. **latitude** – Latitude location of the property

8. **longitude** – Longitude location of the property

9. **room_type** – Type of room (Entire home/apt, Private room, Shared room)

10. **price** – Price per night in dollars

11. **minimum_nights** – Minimum number of nights required to book

12. **number_of_reviews** – Total number of reviews for the listing

13. **last_review** – Date of the most recent review

14. **reviews_per_month** – Average number of reviews per month

15. **calculated_host_listings_count** – Number of listings owned by the host

16. **availability_365** – Number of days the listing is available in a year

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,root_mean_squared_error
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,StackingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()

# fit scaler ONLY on training data
x_train_scaled = scaler.fit_transform(x_train)

# transform test data
x_test_scaled = scaler.transform(x_test)
```

```python
models = {
    "LinearRegression": LinearRegression(),
    "RandomForestRegressor": RandomForestRegressor(),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "SVM": SVR(),
    
}  
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd

# Scaling
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

results = {
    "Model": [],
    "MSE": [],
    "MAE": [],
    "R2": [],
    "RMSE": []
}

for model_name, model in models.items():

    # Linear & SVM → scaled data
    if model_name in ["LinearRegression", "SVM"]:
        model.fit(x_train_scaled, y_train)
        y_pred = model.predict(x_test_scaled)

    # Tree models → original data
    else:
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)

    results["Model"].append(model_name)
    results["MSE"].append(mse)
    results["MAE"].append(mae)
    results["R2"].append(r2)
    results["RMSE"].append(rmse)

results_df = pd.DataFrame(results).sort_values(by="MSE")
results_df.style.background_gradient("Blues")
```

# Tuning

```python
models = {
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {
            "fit_intercept": [True, False]
        }
    },
    

    "RandomForestRegressor": {
        "model": RandomForestRegressor(random_state=42),


        "params": {
            "n_estimators": [100,200],
            "max_depth": [None, 10, 20],
            "min_samples_split": [2, 5]
        }
    },

    "GradientBoostingRegressor": {
        "model": GradientBoostingRegressor(random_state=42),
        "params": {
            "n_estimators": [100, 200],
            "learning_rate": [0.05, 0.1],
            "max_depth": [3, 5]
        }
    },

    "DecisionTreeRegressor": {
        "model": DecisionTreeRegressor(random_state=42),
        "params": {
            "max_depth": [None, 10, 20],
            "min_samples_split": [2, 5]
        }
    },

#     "SVM": {

        
#         "model": SVR(),
#         "params": {
#             "C": [1, 10, 100],
#             "epsilon": [0.1, 0.2],
#             "kernel": ["rbf", "linear"]
#         }
#     }
 }




results = {
    "Model": [],
    "Best_Params": [],
    "MSE": [],
    "MAE": [],
    "RMSE": [],
    "R2": []
}



for model_name, mp in models.items():

    print(f"Tuning {model_name}...")

    grid = GridSearchCV(
        estimator=mp["model"],
        param_grid=mp["params"],
        cv=3,
        scoring="neg_mean_squared_error",
        n_jobs=-1
    )

    grid.fit(x_train, y_train)

    best_model = grid.best_estimator_

    y_pred = best_model.predict(x_test)

    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    results["Model"].append(model_name)
    results["Best_Params"].append(grid.best_params_)
    results["MSE"].append(mse)
    results["MAE"].append(mae)
    results["RMSE"].append(rmse)
    results["R2"].append(r2)



results_df = pd.DataFrame(results)
results_df.sort_values(by="MSE", ascending=True)
```

# Deploy 

```python
import gradio as gr
import pandas as pd
import pickle

# Load trained model
model = pickle.load(open("Airbnb_model.pkl", "rb"))

# Manual encoders (VERY IMPORTANT)
neighbourhood_map = {
    "Manhattan": 0,
    "Brooklyn": 1,
    "Queens": 2,
    "Bronx": 3,
    "Staten Island": 4
}

room_type_map = {
    "Entire home/apt": 0,
    "Private room": 1,
    "Shared room": 2
}

def predict_price(
    neighbourhood_group,
    latitude,
    longitude,
    room_type,
    minimum_nights,
    number_of_reviews,
    calculated_host_listings_count,
    availability_365
):
    # Encode categorical values
    neighbourhood_group = neighbourhood_map[neighbourhood_group]
    room_type = room_type_map[room_type]

    # Create dataframe in SAME order as training
    input_df = pd.DataFrame([[
        neighbourhood_group,
        latitude,
        longitude,
        room_type,
        minimum_nights,
        number_of_reviews,
        calculated_host_listings_count,
        availability_365
    ]], columns=[
        "neighbourhood_group",
        "latitude",
        "longitude",
        "room_type",
        "minimum_nights",
        "number_of_reviews",
        "calculated_host_listings_count",
        "availability_365"
    ])

    prediction = model.predict(input_df)
    price = max(0, prediction[0])

    return f"Predicted Airbnb Price: ${price:.2f}"


app = gr.Interface(
    fn=predict_price,
    inputs=[
        gr.Dropdown(
            ["Manhattan", "Brooklyn", "Queens", "Bronx", "Staten Island"],
            value="Manhattan",
            label="Neighbourhood Group"
        ),
        gr.Number(value=40.7128, label="Latitude"),
        gr.Number(value=-74.0060, label="Longitude"),
        gr.Dropdown(
            ["Entire home/apt", "Private room", "Shared room"],
            value="Entire home/apt",
            label="Room Type"
        ),
        gr.Number(value=3, label="Minimum Nights"),
        gr.Number(value=20, label="Number of Reviews"),
        gr.Number(value=2, label="Host Listings Count"),
        gr.Number(value=180, label="Availability (365 days)")
    ],
    outputs="text",
    title="NYC Airbnb Price Prediction",
    description="Predict Airbnb prices using Machine Learning (Gradio Deployment)"
)

app.launch()
```
